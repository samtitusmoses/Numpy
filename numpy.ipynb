## Min & Max

```python
# Min & Max
print(two_d)
print("Min element",np.min(two_d))
print("Max elemet",np.max(two_d))
```

```python
print(two_d)
print("Min position",np.argmin(two_d))
print("Max Position",np.argmax(two_d))
```

## Quering (Searching)

```python
print(one_d)
print("-----------Position where great than 1 -----------------")
x=np.where(one_d>1)
print(x)
print("=======>  elements greater than 1  <==========")
print(one_d[x])
```

```python
print(two_d)
print("-----------Position where great than 2 -----------------")
x=np.where(two_d>2)
print(x)
print("=======>  elements greater than 2  <==========")
print(two_d[x])
```
